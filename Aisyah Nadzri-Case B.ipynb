```python
# Insert data into MongoDB
protein_col = db['protein_info']
organism_col = db['organism_info']

# Insert documents into respective collections
protein_col.insert_many(protein_info_dict)
organism_col.insert_many(organism_info_dict)
```

```python
# Create a dictionary for quick lookup of organism information by CommonName
organism_lookup = {org['CommonName']: org for org in organism_info_dict}

# Iterate over each protein document and update it with the corresponding organism details
for protein in protein_info_dict:
    organism_name = protein['Organism']
    if organism_name in organism_lookup:
        organism_details = organism_lookup[organism_name].copy()  # Use copy to avoid modifying the original dictionary
        # Remove CommonName from organism details to avoid duplication
        del organism_details['CommonName']
        # Update the protein document with the organism details
        protein_col.update_one(
            {"ProteinID": protein['ProteinID']},
            {"$set": {"OrganismDetails": organism_details}}
        )

print("Protein information updated with organism details.")
```
